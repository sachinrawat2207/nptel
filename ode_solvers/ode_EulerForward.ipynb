1. Solve the ode $\dot{x} = \sin(t)$ with $x(0) = 10$ using Euler's forward methods and find out the value of $x$ at $t=5$. 

```python
import numpy as np
t_i = 0
t_f = 5

dt = 0.001

n = int((t_f-t_i)/dt) +1

x = np.zeros(n)
t = np.linspace(t_i, t_f, n)

def f(x,t):
    return np.sin(t)
    
x[0] = 10
for i in range(n-1):
    x[i+1] = x[i] + dt*f(x[i],t[i])
    
print(x)
```

Exact solution of the ode is $x = -\cos(t) + 11$

```python
from matplotlib import pyplot as plt

y = -np.cos(t) + 11
plt.plot(t,np.abs(x-y))
# plt.plot(t,y,'--')
```
