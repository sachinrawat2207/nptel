## RK2 Method
For ODE $\displaystyle\frac{dx}{dt}=f(x,t)$

Step 1: $x^{n+1/2} = x^n+\Delta t f(x^{n}, t^{n})$

Step 2: $x^{n+1}=x^{n}+\Delta t f(x^{n+1/2}, t^{n+1/2})$

1. Solve the ode $\dot{x} = \sin(t)$ with $x(0) = 10$ using RK2 methods and find out the value of $x$ at $t=5$. 

```python
import numpy as np

t_i = 0
t_f = 5
dt = 0.001
n = int((t_f-t_i)/dt +1)
x = np.zeros(n)
t = np.linspace(t_i, t_f, n)
x[0] = 10

def f(x,t):
    return np.sin(t)

for i in range(n-1):
    x[i+1] = x[i]+dt*f(x[i], t[i])
    x[i+1] = x[i]+dt*f(x[i+1], t[i]+dt/2)

print(t[-1],x[-1])
```

Exact solution of the ode is $x = -\cos(t) + 11$

```python
from matplotlib import pyplot as plt
y = -np.cos(t) + 11
error = np.abs(y-x)
plt.plot(t, y, label = 'Exact')
plt.plot(t, x, '--' , label = 'RK2')
plt.xlim([t[0], t[-1]])
# plt.ylim([x[0],x[-1]])
plt.legend()
plt.show()
```
