1. Solve the ode $\dot{x} = x^{-1}$ with $x(0) = 1$ using Euler's backward methods and find out the value of $x$ at $t=10$. 

```python
import numpy as np
t_i = 0
t_f = 10

dt = 0.001

n = int((t_f-t_i)/dt) +1

x = np.zeros(n)
t = np.linspace(t_i, t_f, n)
x[0] = 1
for i in range(n-1):
    x[i+1] = (x[i]+np.sqrt(x[i]**2+4*dt))/2
print(x)
```

2. The exact solution of the ode is $x = \pm\sqrt{2t+1}$

```python
from matplotlib import pyplot as plt
y = np.sqrt(2*t+1)

plt.plot(t, np.abs(x-y))
# plt.plot(t, y, label = 'Actual')
# plt.plot(t, x, '--', label = 'EB')
# plt.xlabel('t')
# plt.ylabel('x')
# plt.legend()
plt.show()
```
