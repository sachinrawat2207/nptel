1. Solve the ode $\dot{x} = \displaystyle\frac{x}{t} + 2t$ with $x(5) = 0$ using Predictor-corrector method and find out the value of $x$ at $t=10$. 

Step 1: $x_p^{n+1} = x^n+\Delta t f(x^n,t^n) $ 

Step 2: $x^{n+1} = x^n+\Delta t [f(x^n,t^n)+f(x_p^{n+1},t^{n+1})]/2 $

```python
import numpy as np

t_i = 5
t_f = 10

dt = 0.001

n = int((t_f - t_i)/dt) + 1

t = np.linspace(t_i, t_f, n)
x = np.zeros(n)

x[0] = 0

def f(x, t):
    return x/t + 2*t

for i in range(n-1):
    x[i+1] = x[i] + dt*f(x[i], t[i])
    x[i+1] = x[i] + dt/2*(f(x[i], t[i])+f(x[i+1], t[i+1]))

print(t[-1],x[-1])
```

2. The exact solution of the differential equation is $ x = 2t^2-10t$

```python
from matplotlib import pyplot as plt
y = 2*t**2-10*t
error = np.abs(y-x)
plt.plot(t,error)
# plt.plot(t,y, '--', label = 'Exact')
# plt.xlabel('t')
# plt.ylabel('x')
# plt.xlim([5,10])
# plt.ylim([x[0],x[-1]])
# plt.legend()
plt.show()
```
